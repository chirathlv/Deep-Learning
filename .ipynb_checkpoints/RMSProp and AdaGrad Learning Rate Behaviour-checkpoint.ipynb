```python
# import packages
import sys
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# load the dataset
df = pd.read_csv("../../DataSet/mnist_train.csv")
```

```python
# input and output features
X = df.iloc[:,1:].to_numpy() / 255.0 # Normalizing between 0 to 1
Y = df.iloc[:,0].to_numpy()
```

```python
# split into train and test data
Xtrain, Xtest, Ytrain, Ytest = X[:-1000], X[-1000:], Y[:-1000], Y[-1000:]
```

```python
# Shuffle the training data
Xtrain, Ytrain = shuffle(Xtrain, Ytrain)
```

```python
# Show PIC func
def showPIC(x,y):
    plt.figure(figsize=(2,2))
    plt.title("Digit: "+str(y))
    plt.imshow(x.reshape(28,28))
    plt.show()
```

```python
# Testing a PIC
pic_id = 1
showPIC(Xtrain[pic_id], Ytrain[pic_id])
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def tanh(z):
    return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))
```

```python
def relu(z):
    return z*(z>0)
```

```python
def softmax(z):
    return np.exp(z) / np.exp(z).sum(axis=1, keepdims=True)
```

```python
def oneHotEncode(y):
    T = np.zeros((y.shape[0], np.unique(y).shape[0]))
    
    for i,val in enumerate(y):
        T[i,val] = 1
        
    return T
```

```python
def nonlinearDr(z,activation):
    if activation=="sigmoid":
        return z*(1-z)
    if activation=="tanh":
        return (1-z**2)
    if activation=="relu":
        return (z>0)
```

```python
def cost(t,y):
    return -np.sum(t*np.log(y))
```

```python
def forward(X,T,hidden,activation,W,B,A,Vw,Vb,cache_W,cache_B,i,seed):
    D = X.shape[1]
    K = T.shape[1]
    L = len(hidden)
    
    for l in range(L+1):
        if l==0:
            if i==0:
                W[l] = np.random.randn(D,hidden[l])
                B[l] = np.random.randn(hidden[l])
                Vw[l] = 0
                Vb[l] = 0
                cache_W[l] = 0
                cache_B[l] = 0
            A[l] = activation(np.dot(X,W[l])+B[l])
            
        elif l==L:
            if i==0:
                W[l] = np.random.randn(hidden[l-1],K)
                B[l] = np.random.randn(K)
                Vw[l] = 0
                Vb[l] = 0
                cache_W[l] = 0
                cache_B[l] = 0
            A[l] = softmax(np.dot(A[l-1],W[l])+B[l])
            Y = A[l]
            
        else:
            if i==0:
                W[l] = np.random.randn(hidden[l-1],hidden[l])
                B[l] = np.random.randn(hidden[l])
                Vw[l] = 0
                Vb[l] = 0
                cache_W[l] = 0
                cache_B[l] = 0
            A[l] = activation(np.dot(A[l-1],W[l])+B[l])
    
    return W,B,A,Vw,Vb,cache_W,cache_B,Y
```

```python
def backprop(X,T,Y,hidden,activation,lr,momentum,Nesterov,AdaGrad,RMSProp,decay,cache_W,cache_B,W,B,A,Vw,Vb):
    L = len(hidden)
    Err,Grad = {},{}
    
    for l in range(L,-1,-1):
        if l==L:
            Err[l] = T-Y
            Grad[l] = Err[l]
            
        else:
            Err[l] = np.dot(Grad[l+1],W[l+1].T)
            Grad[l] = Err[l]*nonlinearDr(A[l],activation.__name__)
            
    for l in range(L+1):
        if AdaGrad:
            if l==0:
                gradW = np.dot(X.T,Grad[l])
                gradB = Grad[l].sum(axis=0)
                cache_W[l] += gradW ** 2
                cache_B[l] += gradB ** 2
                lr_W = lr / (np.sqrt(cache_W[l]) + 1e-8)
                lr_B = lr / (np.sqrt(cache_B[l]) + 1e-8)
                W[l] += lr_W * gradW 
                B[l] += lr_B * gradB
            else:
                gradW = np.dot(A[l-1].T, Grad[l])
                gradB = Grad[l].sum(axis=0)
                cache_W[l] += gradW ** 2
                cache_B[l] += gradB ** 2
                lr_W = lr / (np.sqrt(cache_W[l]) + 1e-8)
                lr_B = lr / (np.sqrt(cache_B[l]) + 1e-8)
                W[l] += lr_W * gradW 
                B[l] += lr_B * gradB
                
        elif RMSProp:
            if l==0:
                gradW = np.dot(X.T,Grad[l])
                gradB = Grad[l].sum(axis=0)
                cache_W[l] = cache_W[l] * decay + (1 - decay) * gradW ** 2
                cache_B[l] = cache_B[l] * decay + (1 - decay) * gradB ** 2
                lr_W = lr / (np.sqrt(cache_W[l]) + 1e-8)
                lr_B = lr / (np.sqrt(cache_B[l]) + 1e-8)
                W[l] += lr_W * gradW
                B[l] += lr_B * gradB
            else:
                gradW = np.dot(A[l-1].T, Grad[l])
                gradB = Grad[l].sum(axis=0)
                cache_W[l] = cache_W[l] * decay + (1 - decay) * gradW ** 2
                cache_B[l] = cache_B[l] * decay + (1 - decay) * gradB ** 2
                lr_W = lr / (np.sqrt(cache_W[l]) + 1e-8)
                lr_B = lr / (np.sqrt(cache_B[l]) + 1e-8)
                W[l] += lr_W * gradW
                B[l] += lr_B * gradB
                
        elif Nesterov:
            if l==0:
                Vw[l] = Vw[l] * momentum + lr * np.dot(X.T,Grad[l])
                Vb[l] = Vb[l] * momentum + lr * Grad[l].sum(axis=0)
                W[l] += Vw[l] * momentum + lr * np.dot(X.T,Grad[l])
                B[l] += Vb[l] * momentum + lr * Grad[l].sum(axis=0)
            else:
                Vw[l] = Vw[l] * momentum + lr * np.dot(A[l-1].T, Grad[l])
                Vb[l] = Vb[l] * momentum + lr * Grad[l].sum(axis=0)
                W[l] += Vw[l] * momentum + lr * np.dot(A[l-1].T, Grad[l])
                B[l] += Vb[l] * momentum + lr * Grad[l].sum(axis=0)
                
        else:
            if l==0:
                Vw[l] = Vw[l] * momentum + lr * np.dot(X.T,Grad[l])
                Vb[l] = Vb[l] * momentum + lr * Grad[l].sum(axis=0)
                W[l] += Vw[l]
                B[l] += Vb[l]
            else:
                Vw[l] = Vw[l] * momentum + lr * np.dot(A[l-1].T, Grad[l])
                Vb[l] = Vb[l] * momentum + lr * Grad[l].sum(axis=0)
                W[l] += Vw[l]
                B[l] += Vb[l]
            
    return W,B,Vw,Vb,cache_W,cache_B,lr_W,lr_B
```

```python
def train(X,Ytrain,Xtest,Ytest,hidden,activation,iterations,lr,momentum,Nesterov,AdaGrad,RMSProp,decay,batch_size,seed):
    N = X.shape[0]
    n_batches = N // batch_size
    
    T = oneHotEncode(Ytrain)
    Ttest = oneHotEncode(Ytest)
    W,B,A,Vw,Vb,cache_W,cache_B = {},{},{},{},{},{},{}
    Ctrain, Ctest, lr_W_all, lr_B_all = [],[],[],[]
    
    for i in range(iterations):
        for j in range(n_batches):
            start = batch_size*j
            end = batch_size*(j+1)
            
            W,B,A,Vw,Vb,cache_W,cache_B,Y = forward(X[start:end],T[start:end],hidden,activation,W,B,A,Vw,Vb,cache_W,cache_B,i,seed)
            W,B,Vw,Vb,cache_W,cache_B,lr_W,lr_B = backprop(X[start:end],T[start:end],Y,hidden,activation,lr,momentum,Nesterov,AdaGrad,RMSProp,decay,cache_W,cache_B,W,B,A,Vw,Vb)
            _,_,_,_,_,_,_,Y_test = forward(Xtest,Ttest,hidden,activation,W,B,A,Vw,Vb,cache_W,cache_B,2,seed)

            trCost = cost(T[start:end],Y)
            teCost = cost(Ttest,Y_test)

            Ctrain.append(trCost)
            Ctest.append(teCost)
            
            lr_W_all.append(lr_W)
            lr_B_all.append(lr_B)

            sys.stdout.write("\rIterations: {} | Batch: {} | Training Cost: {} | Test Cost: {}".format(
                i+1, j+1, trCost, teCost))
    
    _,_,_,_,_,_,_,Y_train = forward(Xtrain,Xtest,hidden,activation,W,B,A,Vw,Vb,cache_W,cache_B,2,seed) 
    
    return Ctrain, Ctest, W, B, Vw, Vb, lr_W_all, lr_B_all, Y_train, Y_test
```

```python
# Mini Batch Gradient Descent with AdaGrad
hidden = [100]
activation = sigmoid
iterations = 100
lr = 0.1 # Increased the Learning Rate from 0.0001
momentum = 0.95
Nesterov = False # Turned this OFF
AdaGrad = True # Turned this ON
RMSProp = False # Turned this OFF
decay = 0
batch_size = 10000 # Batch Sample size
seed = 1

Ctrain_batch_Ada, Ctest_batch_Ada, W, B, Vw, Vb, lr_W_ada, lr_B_ada, Y, Y_test = train(Xtrain,Ytrain,Xtest,Ytest,hidden,
                                                                                         activation,iterations,lr,
                                                                                         momentum,
                                                                                         Nesterov,
                                                                                         AdaGrad,
                                                                                         RMSProp,
                                                                                         decay,
                                                                                         batch_size,
                                                                                         seed)
```

```python
# Mini Batch Gradient Descent with RMSProp
hidden = [100]
activation = sigmoid
iterations = 100
lr = 0.01 # Increased the Learning Rate from 0.0001
momentum = 0
Nesterov = False # Turned this OFF
AdaGrad = False # Turned this OFF
RMSProp = True # Turned this ON
decay = 0.999 # Set the value
batch_size = 10000 # Batch Sample size
seed = 1

Ctrain_batch_RMS, Ctest_batch_RMS, W, B, Vw, Vb, lr_W_rms, lr_B_rms, Y_train, Y_test = train(Xtrain,Ytrain,Xtest,Ytest,hidden,activation,iterations,lr,
                                                         momentum,
                                                         Nesterov,
                                                         AdaGrad,
                                                         RMSProp,
                                                         decay,
                                                         batch_size,
                                                         seed)
```

```python
Ada_Lr = []
RMS_Lr = []
for i,_ in enumerate(lr_W_ada):
    # i <- Goes upto Num_iterations 100*5 = 500
    # lr_W_ada[i][0] = 1st Node in the hidden layer (100 nodes in total)
    # lr_W_ada[i][0][0] = Weight between connecting 1st Node in hidden and 1st node in output layer 
    Ada_Lr.append(lr_W_ada[i][0][0])
    
for i,_ in enumerate(lr_W_rms):
    # i <- Goes upto Num_iterations 100*5 = 500
    # lr_W_ada[i][0] = 1st Node in the hidden layer (100 nodes in total)
    # lr_W_ada[i][0][0] = Weight between connecting 1st Node in hidden and 1st node in output layer 
    RMS_Lr.append(lr_W_rms[i][0][0]) 
    
plt.figure(figsize=(6,6))
plt.plot(Ada_Lr, label="AdaGrad Learning Rate")
plt.plot(RMS_Lr, label="RMSProp Learning Rate")
plt.legend()
plt.show()
```

Here we can see Learning Rate is Aggresivley become smaller With AdaGrad. So there is little learning according to Weight update Rule over time because weight change is very small since learning rate has become very small 

However, RMSProp restrict learning rate aggresiveness using decay parameter. So, weight update is still happening and learning is continuing with slow rate (less aggresive over time)
