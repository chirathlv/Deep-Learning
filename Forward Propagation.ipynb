# Forward Propagation

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create 3 Gaussian Clouds with 500 samples each

N = 500 # N.samples
D = 2 # N.Features
H = 3 # N.Hidden Units
K = 3 # N.classes

X1 = np.random.randn(N,D) + np.array([0, -2])
X2 = np.random.randn(N,D) + np.array([2, 2])
X3 = np.random.randn(N,D) + np.array([-2, 2])

X = np.vstack([X1, X2, X3])
Y = np.array([0]*N + [1]*N + [2]*N)
```

```python
print(X.shape, Y.shape)
```

```python
# Visualize
plt.figure(figsize=(8,5))
plt.scatter(X[:,0], X[:,1], c=Y, s=100, alpha=0.5)
plt.plot()
plt.show()
```

```python
# Weights initialization
W1 = np.random.randn(D,H)
b1 = np.random.randn(H)

W2 = np.random.randn(H,K)
b2 = np.random.randn(K)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def softmax(z):
    return np.exp(z) / np.sum(np.exp(z))
```

```python
def prediction(output):
    return np.argmax(output, axis=1)
```

```python
def accuracy(y, yhat):
    return np.mean(y==yhat)
```

```python
# Input to Hidden layer
a1 = sigmoid(np.dot(X,W1)+b1)

# Hidden to output layer
a2 = np.dot(a1, W2) + b2

# Apply softmax
output = softmax(a2)
```

```python
# prediction
Yhat = prediction(output)

# accuracy
results = accuracy(Y, Yhat)
```

```python
print("Accuracy: ", str(results*100))
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.title("Prediction", fontsize=15)
plt.scatter(X[:,0], X[:,1], c=Yhat, s=100, alpha=0.5)
plt.subplot(1,2,2)
plt.scatter(X[:,0], X[:,1], c=Y, s=100, alpha=0.5)
plt.title("Original", fontsize=15)
plt.show()
```
